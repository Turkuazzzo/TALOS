**Neden NumPy**

```python
import time

z1 = time.time()

liste = []
for a in range(1000000):
    kuvvet = a ** 100
    liste.append(kuvvet)

z2 = time.time()

print(z2-z1)

liste2 = np.arange(1000000)**100

z3 = time.time() # normal basic python'dan 100 kat daha hızlı

print(z3-z2)
```

**eye(), diag(), identity() Metodları**

```python
print(np.eye(3,5,k=1)) # birim matris oluşturur nxm tipinde ve k = birim matrisin ne kadar kaydırılacağı

print(np.diag([1,1,1,1]))

print(np.identity(5))
```

**Mantıksal Operatörler ve Filtreleme**

```python
matris = np.random.randint(0,10,(5,5))

matris
```

```python
matris>5
```

```python
matris[matris>5]
```

```python
matris[matris==5]
```

```python
matris[matris==5].size
```

```python
sifir = matris == 0
cift = matris%2 == 0
bir = matris == 1

print(matris[cift])
print(matris[sifir])
print(matris[bir])
```
